## Clean up data
Only want USA Racers, 50K or 50Mi, 2020

```python
# Combine 50km, 50mi
df[df['Event distance/length'].isin(['50km' , '50mi'])]
```

```python
# Add year 2020
df[(df['Event distance/length'].isin(['50km' , '50mi'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# Add USA
df2 = df[(df['Event distance/length'].isin(['50km' , '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

```python
# Remove USA from Event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(5)
```

```python
# Clean up athlete age
df2['Athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head(5)
```

```python
# Remove h from Athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
# Drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
df2.head(5)
```

```python
df2.shape
```

```python
# Clean up null values
df2.isna().sum()
```

```python
df2[df2['Athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
df2.shape
```

```python
# check for dupes
df2[df2.duplicated() == True]
```

```python
# Reset index
df2.reset_index(drop=True)
```

```python
# Fix types
df2.dtypes
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int) 
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float) 
df2.dtypes
```

```python
df2.head(5)
```

```python
# Rename columns
df2 = df2.rename(columns= {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name' : 'race_name',
    'Event distance/length' : 'race_lenght',
    'Event number of finishers' : 'race_number_of_finishers',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'athlete_gender',
    'Athlete average speed' : 'athlete_average_speed',
   'Athlete ID' : 'athlete_id',
   'Athlete_age' : 'athlete_age'
})
df2.head()
```

```python
# Reorder columns
df3 = df2[['race_day', 'race_name', 'race_lenght', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
df3.head()
```

## Charts and Graphs

```python
# Difference in speed for the 50k, 50mi male to female
df3.groupby(['race_lenght', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age groups are the best in the 50mi race (20 + races min)
df3.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

```python
# What age groups are the worst in the 50mi race (20 + races min)
df3.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count>19').head(15)
```

Seasons for the data -> slower in summer than winter?

spring 3-5
summer 6-8
fall 9-11
winter 12-2

split between two decimals

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
df3.head(30)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
df3.head(25)
```

```python
df3.query('race_month == 1').head(15)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

```python
df3.query('race_lenght == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
